```python
import pandas as pd
df = pd.read_csv('fatalities_isr_pse_conflict_2000_to_2023.csv')
df.head()
```

```python
df.info()
```

```python
print('Number of rows and column :')
print({df.shape[0], df.shape[1]})
print('Number of duplicate entries in the dataset are : ')
print(df.duplicated().sum())
print('Number missing values in the dataset are : ')
print(sum(df.isna().sum()))
```

# Dealing with Duplicate

```python
df.drop_duplicates(keep='first',inplace=True)
print('Number of duplicate entries in the dataset are : ')
print(df.duplicated().sum())
```

# Change data type for Date 

```python
df['date_of_event'] = pd.to_datetime(df['date_of_event'])
df['date_of_death'] = pd.to_datetime(df['date_of_death'])

df.info()
```

# Seperate Date 

```python
df['e_year'] = pd.to_datetime(df['date_of_event']).dt.year
df['e_month'] = pd.to_datetime(df['date_of_event']).dt.month
df['e_day'] = pd.to_datetime(df['date_of_event']).dt.day

df['d_year'] = pd.to_datetime(df['date_of_death']).dt.year
df['d_month'] = pd.to_datetime(df['date_of_death']).dt.month
df['d_day'] = pd.to_datetime(df['date_of_death']).dt.day

print(df[['date_of_event', 'e_year', 'e_month', 'e_day']].head(5))
print(df[['date_of_death', 'd_year', 'd_month', 'd_day']].head(5))
```

# Describe

```python
df.describe(include='all').transpose()
```

# Handling Null Values

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df['gender'].fillna(df['gender'].mode()[0], inplace=True)
```

```python
df.dropna(subset=['took_part_in_the_hostilities'], inplace=True)
```

```python
df['place_of_residence'].fillna('Unknown', inplace=True)
df['place_of_residence_district'].fillna('Unknown', inplace=True)
```

```python
df['notes'].fillna('Not Specified', inplace=True)
```

```python
df['type_of_injury'].fillna('Not Specified', inplace=True)
df['ammunition'].fillna('Not Specified', inplace=True)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df.to_csv('Fatalities-Israel-Palestinian-Conflict.csv')
```
